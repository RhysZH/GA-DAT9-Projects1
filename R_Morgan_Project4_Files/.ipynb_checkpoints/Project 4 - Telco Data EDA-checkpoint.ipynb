```python
#function to check correlation of two features
def corr_cols(x,y):
    print(df[x].corr(df[y]))

corr_cols('contacts_during_campaign','last_comm_duration')
```

```python
#function to get standard deviation of feature
def bank_std(col1):
    print(df[col1].std())
    
bank_std('contacts_during_campaign')
```

```python
pivot_items = pd.pivot_table(df, values='num_prev_contacts', index=['education'], aggfunc=np.sum)
pivot_items
```

```python
conversion_map = {' False.':0,' True.':1}
yn_map = {' no':'0',' yes':'1'}
df['international_plan_n'] = df['international_plan'].map(yn_map)
df['voice_mail_plan_n'] = df['voice_mail_plan'].map(yn_map)
df['area_code'] = df.area_code.astype(str)
df['international_plan_n'] = df.international_plan_n.astype(int)
df['voice_mail_plan_n'] = df.voice_mail_plan_n.astype(int)
df = pd.get_dummies(df, columns=['area_code'])
df['churn_n'] = df['churn'].map(conversion_map)
df.drop(['state','phone_number','international_plan','voice_mail_plan','churn','total_international_minutes','total_day_minutes','total_eve_minutes','total_night_minutes'],axis=1, inplace=True)
```
